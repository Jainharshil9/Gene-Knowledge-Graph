```python
import json
import pandas as pd

# Load the CSV file
nodes_df = pd.read_csv('nodes_for_neo4j.csv')

# Initialize an empty list for the node schema
node_schema = []

# Populate the node schema based on the data from the DataFrame
for index, row in nodes_df.iterrows():
    node = {
        "id": row['id'],
        "name": row['name'],
        "ncbi_id": row['ncbi_id']
    }
    node_schema.append(node)

# Example of what to include in 'rest' if needed
rest = {}

# Construct the full schema dictionary
schema = {
    "nodes": node_schema,
    "edges": [],  # Assuming edge_schema creation is handled elsewhere
    **rest
}

# Write the schema to a JSON file
with open("../schema.json", 'w') as o:
    o.write(json.dumps(schema, indent=4))

print("Schema written successfully with", len(node_schema), "nodes.")
```

```python
import json
import pandas as pd

# Load the CSV file
edges_df = pd.read_csv('edges_for_neo4j.csv')

# Initialize an empty list for the edge schema
edge_schema = []

# Populate the edge schema based on the data from the DataFrame
for index, row in edges_df.iterrows():
    edge = {
        "source": row['source'],
        "target": row['target'],
        "score": row['normalized_score'],
        "type": row['interaction_type']
    }
    edge_schema.append(edge)

# Example of what to include in 'rest' if needed
rest = {}

# Construct the full schema dictionary
schema = {
    "nodes": [],  # Assuming node_schema creation is handled elsewhere
    "edges": edge_schema,
    **rest
}

# Write the schema to a JSON file
with open("../schema.json", 'w') as o:
    o.write(json.dumps(schema, indent=4))

print("Schema written successfully with", len(edge_schema), "edges.")
```
